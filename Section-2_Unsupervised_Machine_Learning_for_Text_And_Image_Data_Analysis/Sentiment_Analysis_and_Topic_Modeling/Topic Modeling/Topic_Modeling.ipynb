### Text Preprocessing

```python
import pandas as pd
```

```python
cust_data = pd.read_csv('/content/complaint_data.csv')
```

```python
cust_data.head()
```

```python
# Filtering the first 50000 records
data = cust_data.iloc[:50000]
print("\nNumber of records:",len(data))
```

```python
# Displaying names of the columns
print("Columns in dataset:\n",data.columns)
```

```python
# Filtering 'issue' aolumn for topic modeling ,removing others columns
data = data.drop(columns=['date_received', 'product', 'sub_product','sub_issue',
       'consumer_complaint_narrative', 'company_public_response', 'company',
       'state', 'zip_code', 'tags', 'consumer_consent_provided',
       'submitted_via', 'date_sent_to_company', 'company_response_to_consumer',
       'timely_response', 'consumer_disputed', 'complaint_id'])
```

```python
# Displaying the new dataset
print("\nFirst few recorrds:\n",data.head())
```

```python
# Transformation of data with spacy configuration
from spacy.lang.en import English
import spacy
nlp = spacy.load('en_core_web_sm')
```

```python
data['issue_nlp'] = None
```

```python
# Data tokenization
nlp = spacy.load('en_core_web_sm')
for i in range(0,len(data)):
  data['issue_nlp'][i] = nlp(data['issue'][i])
```

```python
# Removal of stop words,punctuation marks,numbers,url an spaces
data['claen_data'] = None
for i in range(0,len(data)):
  data['claen_data'][i] = [token for token in data['issue_nlp'][i] if not token.is_stop and not token.is_punct and not token.like_num and not token.like_url and not token.is_space]
```

```python
# Lemmatization of data
data['data_lemma'] = None
for i in range(0,len(data)):
  data['data_lemma'][i] = [token.lemma_ for token in data['claen_data'][i]]
```

```python
# Displaying the dataset
print("\nFirst few records:\n",data.head(2))
```

```python
data.head(2)
```

# Document Matrix and Bag of words

```python
import gensim
import gensim.corpora as corpora
```

```python
# Creating a dictionary
id2word = gensim.corpora.Dictionary(data['data_lemma'])
```

```python
# Displaying the dictionary
print("Details of key-value pairs in dictionary")
for i in id2word.items():
  print(i)
```

```python
# Displaying document matrix and bag of words
texts = data['data_lemma']
doc_matrix = [id2word.doc2bow(text) for text in texts]
print("\nDocument matrix:\n",doc_matrix)
[[(id2word[id],freq) for id,freq in cp] for cp in doc_matrix]
```

# LDA  Model Development and Evaluation

```python
doc_matrix
```

LDA - Latent Dirichlet Allocation is one of the most popular topic modeling methods that help to find relevant topics for the document,based on the unique words and their occurrances.

```python
# Developing LDA model
lda_model = gensim.models.ldamodel.LdaModel(corpus=doc_matrix,
                                            id2word=id2word,
                                            num_topics=2,
                                            random_state=100,
                                            update_every=1,
                                            chunksize=100,
                                            passes=10,
                                            alpha='auto',
                                            per_word_topics=True)
# Printing the topics
Topics = lda_model.print_topics()
Topics
```

```python
# Evaluating the model
# Compute and display perplexity of the model
print('\nPerplexity: ', lda_model.log_perplexity(doc_matrix))
```

* Perplexity is a measure of how good the model is.
* Thr lower the perplexity,the better is the model.

```python
# Compute and display coherence score
from gensim.models import CoherenceModel
coherence_model_lda = CoherenceModel(model=lda_model, texts=data['data_lemma'], dictionary=id2word, coherence='c_v')
coherence_lda = coherence_model_lda.get_coherence()
print('\nCoherence Score: ', coherence_lda)
```

# Assigning Topics to Data

```python
# Creating a list of topics
topic_list = []
for i in Topics:
  topic_list.append(list(i))
```

```python
# Assigning names to the topic
topic_list[0][0] = 'Incorrect Report'
topic_list[1][0] = 'Credit Problem'
topic_list
```

```python
# Determinig the topic for each record
data['topic'] = None
```

```python
# Alloting the topic based on the probabilities
for i in range(0,len(data)):
  doc = lda_model.get_document_topics(doc_matrix[i])
  score = []
  for value in doc:
    score.append(value[1])
  count = 0
  for num in score:
    if num == max(score):
      data['topic'][i] = topic_list[count][0]
    count += 1
```

```python
# Determing the number of records having different topics
print(data['topic'].value_counts())
# Displaying the dataset
print("\nFirst few records:\n",data.head())
```
